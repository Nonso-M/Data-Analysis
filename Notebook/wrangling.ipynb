## 2. - Which sellers, if any, should be delisted from this platform?
Two things can make a seller to be delisted;
- He hasn't made any sales  i.e. his id only apprears in the cancelled orders
- Revenue Generated is below a defined threshold

### Assuumptions
- Therer is no data from previous years i.e The company has only operated in the year 2018
- A transaction is successfully completed as long as the status is not saying 'canceled'



```python
df_orders.head()
```

```python
# Extracting unique sellers ids whose orders was successful
sellers_1 = set(df_orders.query('order_status!= "canceled"')["seller_id"].unique())
print(len(sellers_1))

# Extracting sellers ids whose orders weren't successful
sellers_2 = set(df_orders.query('order_status == "canceled"')["seller_id"].unique())
print(len(sellers_2))
```

```python
#Finding Sellers who haven't had a single sucessfull transaction
sellers_flush = sellers_2 - sellers_1
```

```python
#Changing data structures from set to list to use to query the dataframe
sellers_list = list(sellers_flush)

df_sellers = df_orders.query('seller_id in @sellers_list')
```

The Sellers to be delisted can be seen from the output of the code cell below

```python
df_sellers_remove = df_sellers.seller_id.value_counts().reset_index()

#Renaming the columns
df_sellers_remove.columns = ['sellers_id', 'No_of_failed_transactions']
df_sellers_remove 
```
